```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("obs_vs_pred.csv")
df_filtered = df[df['Observed'] <0.99]
```

```python
# Set up a large figure
plt.figure(figsize=(20, 10))

# Create scatterplot for Observed vs Predicted
sns.scatterplot(data=df_filtered, x="Observed", y="Predicted", alpha=0.8, color='darkmagenta', s = 25, markers='+')

# Add confidence intervals as error bars (ymin=Lower, ymax=Upper)
plt.errorbar(df_filtered['Observed'], df_filtered['Predicted'], 
             yerr=[df_filtered['Predicted'] - df_filtered['Lower'], df_filtered['Upper'] - df_filtered['Predicted']], 
             fmt='+', color='skyblue', alpha=0.3, elinewidth=1, capsize=2)

# Add identity line
plt.plot([0, 1], [0, 1], linestyle='--', color='red')

# Labels and title
plt.title("Empirical vs Estimated Prevalence", fontsize=18)
plt.xlabel("Empirical PfPr", fontsize=14)
plt.ylabel("Estimated PfPr (median with 95% CI)", fontsize=14)
#plt.grid(True)

# Show plot
plt.tight_layout()
plt.savefig('/home/incidence_rasters/baseline_old/obs_pred.png')
plt.show()
```
