```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('/home/EVI/evi_features.csv')
```

```python
df.head()
```

```python
print(df.columns.tolist())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# Define predictor variables (feature_0 to feature_1023)
feature_columns = [f'feature_{i}' for i in range(1024)]
X = df[feature_columns]

# Define the outcome variable
y = df['enhanced_vegitation_index']

# Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model using Mean Squared Error
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)  # R-squared value
pearson_corr = np.corrcoef(y_test, y_pred)[0, 1]  # Pearson correlation coefficient
```

```python
# Print results
print(f'Mean Absolute Error: {mae:.4f}')
print(f'R-squared: {r2:.4f}')
print(f'Pearson Correlation: {pearson_corr:.4f}')
```

```python
plt.scatter(y_test, y_pred, color="blue")
plt.ylim(0.0, 0.6)
plt.xlim(0.0, 0.6)
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title("EVI predictions")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# 2000
#train_df = pd.read_csv('/home/EVI/EVI_not2000.csv')
#test_df = pd.read_csv('/home/EVI/EVI_2000.csv')

# 2017
train_df = pd.read_csv('/home/EVI/EVI_not2017.csv')
test_df = pd.read_csv('/home/EVI/EVI_2017.csv')
```

```python
feature_columns = [f'feature_{i}' for i in range(1024)]
X_train = train_df[feature_columns]
y_train = train_df['enhanced_vegitation_index']

X_test = test_df[feature_columns]
y_test = test_df['enhanced_vegitation_index']
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
pearson_corr = np.corrcoef(y_test, y_pred)[0, 1]
```

```python
print(f'Mean Absolute Error: {mae:.4f}')
print(f'R-squared: {r2:.4f}')
print(f'Pearson Correlation: {pearson_corr:.4f}')
```

```python
plt.scatter(y_test, y_pred, color="blue")
plt.ylim(0.0, 0.6)
plt.xlim(0.0, 0.6)
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title("EVI predictions for year 2017 using satellite features")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.show()
```
